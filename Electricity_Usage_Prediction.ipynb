```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("PJM_Load_hourly.csv")
df.head()
```

```python
date_time=pd.to_datetime(df["Datetime"])
date_time.head()
```

```python
#hour column
df["hour"]=date_time.dt.hour
df.head()
```

```python
#day of week
df["day_of_week"]=date_time.dt.dayofweek
df.head()
```

```python
#month extraction
df["month"]=date_time.dt.month
df.head()
```

```python
#is weekend
df["is_weekend"]=df["day_of_week"].apply(lambda x: 1 if x>=5 else 0)
df.head()
```

```python
## holiday
import holidays

holidays=holidays.US()
```

```python
df["is_holiday"]=date_time.dt.date.apply(lambda x:1 if x in holidays else 0)
df.head()
```

```python
df["hour_sin"]=np.sin(2*np.pi*df["hour"] / 24)
df["hour_cos"]=np.cos(2*np.pi*df["hour"] / 24)
df.head()
```

```python
df["month_sin"]=np.sin(2*np.pi*df["month"]/12)
df["month_cos"]=np.cos(2*np.pi*df["month"]/12)
df.head()
```

```python
df.drop(["hour","month","Datetime"],axis=1,inplace=True) 
```

```python
df.head()
```

```python
train_size=int(0.8*len(df))
train=df.iloc[:train_size].copy()
test=df.iloc[train_size:].copy()
```

```python
#calculating lag features for train data 
train["lag24"]=train["PJM_Load_MW"].shift(24)
train["lag168"]=train["PJM_Load_MW"].shift(168)
#calculating rolling features for train data
train["rolling_mean_24"]=train["PJM_Load_MW"].rolling(window=24).mean()
```

```python
#calculating lag features for test data
test["lag24"]=test["PJM_Load_MW"].shift(24)
test["lag168"]=test["PJM_Load_MW"].shift(168)
#calculating rolling features for test data
test["rolling_mean_24"]=test["PJM_Load_MW"].rolling(window=24).mean()
```

```python
train.dropna(inplace=True)
train.head()
```

```python
test.dropna(inplace=True)
test.head(50)
```

```python
features=["day_of_week","is_weekend","is_holiday","hour_sin","hour_cos","month_sin",
         "month_cos","lag24","lag168","rolling_mean_24"]
X_train=train[features]
X_test=test[features]
y_train=train["PJM_Load_MW"]
y_test=test["PJM_Load_MW"]
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='rbf', C=100)
```

```python
svr.fit(X_train,y_train)
```

```python
y_pred=svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("Mean_Squared_Error: ",mean_squared_error(y_test,y_pred))
print("Mean Absolute Error: ",mean_absolute_error(y_test,y_pred))
print("r2_score: ",r2_score(y_test,y_pred))
```

```python
import joblib
joblib.dump(svr,"model.pkl")
joblib.dump(scaler,"scaler.pkl")
```
